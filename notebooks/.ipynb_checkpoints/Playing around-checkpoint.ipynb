```python
tokenizer = Tokenizer()
tokenizer.load_model("../data/tokeniser_data/tokenizer_data.pickle")
```

```python
import csv

lines = []

with open("../data/test.csv", encoding="utf-8") as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        lines.append(row[1])
        lines.append(row[2])
```

```python
for line in lines:
    print("".join(tokenizer.decode([token]) + "|" for token in tokenizer.encode(line)))
```
