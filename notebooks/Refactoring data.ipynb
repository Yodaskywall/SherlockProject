```python
import csv

def get_relevant_info(line, test=False):
    x = "".join([line[1], "|", line[2], "|", line[3]])
    if test:
        return x
    else:
        return [x, line[5]]


with open("../data/train.csv", encoding="utf-8") as file:
    trainCsv = csv.reader(file)
    trainRaw = []
    print(next(trainCsv))
    for line in trainCsv:
        trainRaw.append(get_relevant_info(line))
        
with open("../data/test.csv", encoding="utf-8") as file:
    testCsv = csv.reader(file)
    testRaw = []
    next(testCsv)
    for line in testCsv:
        testRaw.append(get_relevant_info(line, True))
```
